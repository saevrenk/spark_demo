```python
import pyspark
```

```python
from pyspark import SparkContext
```

```python
sc = SparkContext()
```

```python
rdd = sc.textFile("spark_files/cars.csv").zipWithIndex().filter(lambda x: x[1]>1)
rdd.collect()
```

```python
rdd1 = rdd.sample(False, 0.1)
rdd2 = rdd.sample(False, 0.1)
rdd3 = rdd.sample(False, 0.2)
```

```python
rdd_union = sc.union([rdd1, rdd2]).distinct()
rdd_union.collect()
```

```python
rdd3_distinct = rdd3.distinct()
rdd3_distinct.collect()
```

```python
print(rdd_union.count(), rdd3_distinct.count())
```

```python
rddnew = rdd_union.map(lambda l: 1)
length_rdd_union = rddnew.reduce(lambda a,b: a+b)
length_rdd_union
```

```python
rddnew2 = rdd3_distinct.map(lambda l: 1)
length_rdd_union2 = rddnew2.reduce(lambda a,b: a+b)
length_rdd_union2
```

```python
total_weight = rdd.map(lambda x: int(x[0].split(',')[5])).reduce(lambda a,b: a+b)
total_weight
```

```python
total_mpg = rdd.map(lambda x: float(x[0].split(',')[1])).reduce(lambda a,b: a+b)
length = rdd.map(lambda l: 1).reduce(lambda a,b: a+b)
average = total_mpg/length
round(average,2)
```
